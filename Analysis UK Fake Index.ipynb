```python
import numpy as np
import pandas as pd
import csv
import re
```

```python
uk = pd.read_csv('amazon_uk2.csv', header = None, names = ['date', 'star_rating', 'user_name', 'user_type', 'review_text', 'helpful', 'fake_index'])
```

```python
uk2 = pd.read_csv('amazon_uk2cont.csv', header=None, names = ['date', 'star_rating', 'user_name', 'user_type', 'review_text', 'helpful', 'fake_index'])
```

```python
uk2.iloc[66,0] = 'on 14 December 2017'
uk2[uk2['user_name']=='Ibraar']
uk2.iloc[79,0] = 'on 14 November 2017'
```

```python
final_uk = pd.concat((uk, uk2), axis = 0)
```

```python
final_uk['date'] = final_uk['date'].apply(lambda x: x.strip('on '))
final_uk['date'] = pd.to_datetime(final_uk['date'])
final_uk['star_rating'] = final_uk['star_rating'].apply(lambda x: int(x[0]))
```

```python
final_uk['helpful'] = final_uk['helpful'].apply(lambda x: re.sub('One', '1', x))
final_uk['helpful'] = final_uk['helpful'].apply(lambda x: re.sub('[a-zA-Z]', '', x))
final_uk['helpful'] = final_uk['helpful'].apply(lambda x: int(x))
final_uk['fake_index'] = pd.to_numeric(final_uk['fake_index'])

final_uk.index = final_uk.date
final_uk = final_uk.drop('date', axis = 1)
final_uk = final_uk.sort_index()
```

```python
final_uk = final_uk.drop_duplicates()
```
